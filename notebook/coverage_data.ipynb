```python
import pandas as pd
import numpy as np

# Check your data structure
print(df_clean['ANTIGEN'].unique())
print(df_clean['ANTIGEN_DESCRIPTION'].unique())

# Look for dose indicators in antigen descriptions
dose_keywords = ['dose', 'booster', '1st', '2nd', '3rd', '4th', '5th']
for desc in df_clean['ANTIGEN_DESCRIPTION'].unique():
    if any(keyword in str(desc).lower() for keyword in dose_keywords):
        print(f"Found: {desc}")
```

```python
# FIXED VERSION - Only return numeric dose numbers
def extract_dose_number_from_antigen_fixed(antigen, description):
    """Extract dose number from antigen code and description - ONLY NUMBERS"""
    antigen = str(antigen).upper()
    desc = str(description).lower()
    
    # Check antigen codes first - return ONLY NUMBERS
    if antigen == 'DTPCV1' or '1st dose' in desc or ('first' in desc and 'dose' in desc):
        return 1
    elif antigen == 'DTPCV3' or '3rd dose' in desc or ('third' in desc and 'dose' in desc):
        return 3
    elif antigen == 'DIPHCV4' or ('4th' in desc and 'dose' in desc):
        return 4
    elif antigen == 'DIPHCV5' or ('5th' in desc and 'dose' in desc):
        return 5
    elif antigen == 'DIPHCV6' or ('6th' in desc and 'dose' in desc):
        return 6
    elif antigen == 'IPV1' or ('ipv' in desc and '1st' in desc):
        return 1
    elif antigen == 'IPV2' or ('ipv' in desc and '2nd' in desc):
        return 2
    elif antigen == 'MCV1' or ('mcv' in desc and '1st' in desc):
        return 1
    elif antigen == 'MCV2' or ('mcv' in desc and '2nd' in desc):
        return 2
    elif antigen == 'PCV1':
        return 1
    elif antigen == 'PCV2':
        return 2
    elif antigen == 'PCV3':
        return 3
    elif antigen == 'HEPB3':
        return 3
    elif antigen == 'HEPB_BD':
        return 0  # Birth dose - still a number
    elif antigen == 'ROTA1':
        return 1
    # For HPV and others without clear dose numbers, return None or a placeholder
    elif antigen in ['HPV_FEM', 'HPV_MALE']:
        return None  # Not a numbered dose
    elif 'booster' in desc or 'final' in desc or 'last' in desc:
        # These are not sequential doses for drop-off calculation
        return None
    else:
        return None

# Apply the fixed version
df_clean['DOSE_NUMBER'] = df_clean.apply(
    lambda x: extract_dose_number_from_antigen_fixed(x['ANTIGEN'], x['ANTIGEN_DESCRIPTION']), 
    axis=1
)

# Check results - now should only have numbers or None
print("Dose number distribution (fixed):")
print(df_clean['DOSE_NUMBER'].value_counts(dropna=False).sort_index())
print(f"\nTotal rows with numeric dose numbers: {df_clean['DOSE_NUMBER'].notna().sum()}")
```

```python
# Filter for sequential doses (keep all!)
sequential_data = df_clean[df_clean['DOSE_NUMBER'].notna()].copy()
sequential_data['DOSE_NUMBER'] = sequential_data['DOSE_NUMBER'].astype(int)

print(f"Rows with dose numbers: {len(sequential_data)}")
print(f"Rows without dose numbers: {len(df_clean) - len(sequential_data)}")

# Calculate overall drop-off rates
def calculate_dropoff_rates(data):
    """Calculate drop-off between consecutive doses"""
    
    # Get average coverage by dose number
    avg_coverage = data.groupby('DOSE_NUMBER')['COVERAGE'].mean().sort_index()
    
    print("\n" + "=" * 60)
    print("VACCINE DROP-OFF ANALYSIS")
    print("=" * 60)
    
    print(f"\nðŸ“Š Average Coverage by Dose:")
    for dose, coverage in avg_coverage.items():
        print(f"  Dose {dose}: {coverage:.1f}%")
    
    # Calculate drop-off between consecutive doses
    print(f"\nðŸ“‰ Drop-off Rates Between Consecutive Doses:")
    
    doses = sorted(avg_coverage.index)
    for i in range(len(doses) - 1):
        current_dose = doses[i]
        next_dose = doses[i + 1]
        
        current_cov = avg_coverage[current_dose]
        next_cov = avg_coverage[next_dose]
        
        if current_cov > 0 and not pd.isna(next_cov):
            dropoff_pct = ((current_cov - next_cov) / current_cov) * 100
            print(f"  Dose {current_dose} â†’ Dose {next_dose}: {dropoff_pct:.1f}% drop-off")
            print(f"    ({current_cov:.1f}% â†’ {next_cov:.1f}%)")

# Run the analysis
calculate_dropoff_rates(sequential_data)
```

```python
# 1. MOST IMPORTANT: 1st to 3rd dose drop-off (DTP completion)
dose1_avg = sequential_data[sequential_data['DOSE_NUMBER'] == 1]['COVERAGE'].mean()
dose3_avg = sequential_data[sequential_data['DOSE_NUMBER'] == 3]['COVERAGE'].mean()

print("\n" + "=" * 60)
print("KEY FINDING: 1st to 3rd Dose Drop-off")
print("=" * 60)
print(f"\n1st dose average coverage: {dose1_avg:.1f}%")
print(f"3rd dose average coverage: {dose3_avg:.1f}%")

if dose1_avg > 0:
    dropoff_1_to_3 = ((dose1_avg - dose3_avg) / dose1_avg) * 100
    print(f"\nâœ… DROP-OFF RATE (1st â†’ 3rd dose): {dropoff_1_to_3:.1f}%")
    print(f"   Absolute drop: {dose1_avg - dose3_avg:.1f} percentage points")
    
    # Interpretation
    print(f"\nðŸ“‹ ANSWER: {dropoff_1_to_3:.1f}% of people who receive the 1st dose")
    print(f"          do not complete the 3-dose series.")

# 2. Also calculate 1st to 2nd dose drop-off
dose2_avg = sequential_data[sequential_data['DOSE_NUMBER'] == 2]['COVERAGE'].mean()
if dose1_avg > 0 and not pd.isna(dose2_avg):
    dropoff_1_to_2 = ((dose1_avg - dose2_avg) / dose1_avg) * 100
    print(f"\nðŸ“Š Drop-off (1st â†’ 2nd dose): {dropoff_1_to_2:.1f}%")
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Filter for HPV vaccines with gender information
gender_keywords = ['FEM', 'MALE', '_F', '_M']
hpv_gender_data = df_clean[df_clean['ANTIGEN'].str.contains('|'.join(gender_keywords), case=False, na=False)].copy()

print("\n" + "="*70)
print("ANALYSIS: Are vaccination rates different between genders?")
print("="*70)

# Extract gender from antigen codes
def extract_gender(antigen):
    antigen = str(antigen).upper()
    if 'FEM' in antigen or '_F' in antigen:
        return 'Female'
    elif 'MALE' in antigen or '_M' in antigen:
        return 'Male'
    return 'Unknown'

hpv_gender_data['GENDER'] = hpv_gender_data['ANTIGEN'].apply(extract_gender)

print(f"\nðŸ“Š HPV Vaccine Data by Gender:")
print(f"Total HPV vaccine records: {len(hpv_gender_data)}")
print(f"Female records: {len(hpv_gender_data[hpv_gender_data['GENDER'] == 'Female'])}")
print(f"Male records: {len(hpv_gender_data[hpv_gender_data['GENDER'] == 'Male'])}")

# Calculate statistics by gender
gender_stats = hpv_gender_data.groupby('GENDER')['COVERAGE'].agg([
    ('Count', 'count'),
    ('Mean_Coverage', 'mean'),
    ('Median_Coverage', 'median'),
    ('Std_Dev', 'std'),
    ('Min', 'min'),
    ('Max', 'max')
]).round(2)

print("\nðŸ“ˆ Vaccination Coverage Statistics by Gender:")
print(gender_stats)

# Separate data for statistical test
female_data = hpv_gender_data[hpv_gender_data['GENDER'] == 'Female']['COVERAGE'].dropna()
male_data = hpv_gender_data[hpv_gender_data['GENDER'] == 'Male']['COVERAGE'].dropna()

print(f"\nðŸ“‹ Sample sizes:")
print(f"Female: {len(female_data)} coverage measurements")
print(f"Male: {len(male_data)} coverage measurements")

if len(female_data) > 0 and len(male_data) > 0:
    # Perform statistical test
    t_stat, p_value = stats.ttest_ind(female_data, male_data, equal_var=False)
    
    print(f"\nðŸ”¬ Statistical Analysis (Independent T-test):")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.6f}")
    
    # Calculate effect size (Cohen's d)
    pooled_std = np.sqrt((female_data.std()**2 + male_data.std()**2) / 2)
    cohens_d = (female_data.mean() - male_data.mean()) / pooled_std
    
    print(f"Cohen's d (effect size): {cohens_d:.3f}")
    
    # Interpret results
    print(f"\nðŸ“Š INTERPRETATION:")
    
    if p_value < 0.05:
        print(f"âœ… Statistically significant difference (p < 0.05)")
        
        if female_data.mean() > male_data.mean():
            diff = female_data.mean() - male_data.mean()
            print(f"   ðŸ’¡ Females have {diff:.1f}% HIGHER HPV vaccination coverage")
            print(f"   ðŸ’¡ Female coverage: {female_data.mean():.1f}% vs Male: {male_data.mean():.1f}%")
        else:
            diff = male_data.mean() - female_data.mean()
            print(f"   ðŸ’¡ Males have {diff:.1f}% HIGHER HPV vaccination coverage")
            print(f"   ðŸ’¡ Male coverage: {male_data.mean():.1f}% vs Female: {female_data.mean():.1f}%")
        
        # Interpret effect size
        print(f"\nðŸ“ˆ Effect Size Interpretation:")
        if abs(cohens_d) < 0.2:
            print(f"   Small effect (d = {cohens_d:.3f})")
        elif abs(cohens_d) < 0.5:
            print(f"   Medium effect (d = {cohens_d:.3f})")
        else:
            print(f"   Large effect (d = {cohens_d:.3f})")
    else:
        print(f"âŒ No statistically significant difference (p = {p_value:.4f})")
        print(f"   ðŸ’¡ HPV vaccination rates are SIMILAR between genders")
        print(f"   ðŸ’¡ Female: {female_data.mean():.1f}% vs Male: {male_data.mean():.1f}%")
else:
    print("\nâš ï¸ Insufficient data for statistical comparison")

# Additional analysis: Look at trends over time
print("\n" + "="*70)
print("TREND ANALYSIS: Gender Differences Over Time")
print("="*70)

# Group by year and gender
trend_data = hpv_gender_data.groupby(['YEAR', 'GENDER'])['COVERAGE'].mean().unstack()

if not trend_data.empty and 'Female' in trend_data.columns and 'Male' in trend_data.columns:
    print("\nðŸ“… HPV Vaccination Coverage by Year and Gender:")
    print(trend_data.round(1))
    
    # Calculate year-over-year differences
    trend_data['Gender_Gap'] = trend_data['Female'] - trend_data['Male']
    
    print("\nðŸ“Š Gender Gap (Female - Male) by Year:")
    print(trend_data['Gender_Gap'].round(1))
    
    # Check if gap is closing over time
    if len(trend_data) > 1:
        first_year = trend_data.index[0]
        last_year = trend_data.index[-1]
        first_gap = trend_data.loc[first_year, 'Gender_Gap']
        last_gap = trend_data.loc[last_year, 'Gender_Gap']
        
        print(f"\nðŸ“ˆ Trend Analysis:")
        print(f"Gender gap in {first_year}: {first_gap:.1f}%")
        print(f"Gender gap in {last_year}: {last_gap:.1f}%")
        
        if abs(last_gap) < abs(first_gap):
            print(f"âœ… Gender gap is NARROWING over time")
        elif abs(last_gap) > abs(first_gap):
            print(f"âš ï¸ Gender gap is WIDENING over time")
        else:
            print(f"âž¡ï¸ Gender gap is STABLE over time")

# Country-level analysis
print("\n" + "="*70)
print("COUNTRY-LEVEL ANALYSIS: Gender Differences by Country")
print("="*70)

# Get countries with both female and male HPV data
countries_with_both = []

for country in hpv_gender_data['CODE'].unique():
    country_data = hpv_gender_data[hpv_gender_data['CODE'] == country]
    has_female = 'Female' in country_data['GENDER'].values
    has_male = 'Male' in country_data['GENDER'].values
    
    if has_female and has_male:
        countries_with_both.append(country)

if len(countries_with_both) > 0:
    print(f"\nFound {len(countries_with_both)} countries with both female and male HPV data")
    
    # Calculate gender gap for each country
    country_gaps = []
    
    for country in countries_with_both[:10]:  # Show top 10
        country_data = hpv_gender_data[hpv_gender_data['CODE'] == country]
        country_name = country_data['NAME'].iloc[0]
        
        female_avg = country_data[country_data['GENDER'] == 'Female']['COVERAGE'].mean()
        male_avg = country_data[country_data['GENDER'] == 'Male']['COVERAGE'].mean()
        
        if not pd.isna(female_avg) and not pd.isna(male_avg):
            gap = female_avg - male_avg
            country_gaps.append({
                'Country': country_name,
                'Code': country,
                'Female_Coverage': female_avg,
                'Male_Coverage': male_avg,
                'Gender_Gap': gap
            })
    
    # Create DataFrame and sort by absolute gap
    gaps_df = pd.DataFrame(country_gaps)
    gaps_df['Abs_Gap'] = abs(gaps_df['Gender_Gap'])
    gaps_df = gaps_df.sort_values('Abs_Gap', ascending=False)
    
    print("\nðŸ† Countries with Largest Gender Gaps in HPV Vaccination:")
    print(gaps_df[['Country', 'Female_Coverage', 'Male_Coverage', 'Gender_Gap']].round(1).head(10))
    
    # Summary statistics
    print(f"\nðŸ“Š Summary of Gender Gaps across {len(gaps_df)} countries:")
    print(f"Average gender gap: {gaps_df['Gender_Gap'].mean():.1f}%")
    print(f"Maximum female advantage: {gaps_df['Gender_Gap'].max():.1f}%")
    print(f"Maximum male advantage: {gaps_df['Gender_Gap'].min():.1f}%")
    
    # Count countries where each gender has higher coverage
    female_higher = len(gaps_df[gaps_df['Gender_Gap'] > 0])
    male_higher = len(gaps_df[gaps_df['Gender_Gap'] < 0])
    equal = len(gaps_df[gaps_df['Gender_Gap'] == 0])
    
    print(f"\nðŸ“ˆ Distribution of Gender Advantage:")
    print(f"Countries where females have higher coverage: {female_higher}")
    print(f"Countries where males have higher coverage: {male_higher}")
    print(f"Countries with equal coverage: {equal}")

# Final answer
print("\n" + "="*70)
print("âœ… FINAL ANSWER: Are vaccination rates different between genders?")
print("="*70)

if len(female_data) > 0 and len(male_data) > 0:
    if p_value < 0.05:
        if female_data.mean() > male_data.mean():
            print(f"\nYES, vaccination rates ARE different between genders.")
            print(f"ðŸ”¬ Females have SIGNIFICANTLY HIGHER HPV vaccination rates.")
            print(f"ðŸ“Š On average, females have {female_data.mean() - male_data.mean():.1f}% higher coverage.")
            print(f"ðŸ’¡ This suggests gender-based vaccination policies or preferences.")
        else:
            print(f"\nYES, vaccination rates ARE different between genders.")
            print(f"ðŸ”¬ Males have SIGNIFICANTLY HIGHER HPV vaccination rates.")
            print(f"ðŸ“Š On average, males have {male_data.mean() - female_data.mean():.1f}% higher coverage.")
            print(f"ðŸ’¡ This may reflect recent efforts to expand HPV vaccination to boys.")
    else:
        print(f"\nNO, vaccination rates are NOT significantly different between genders.")
        print(f"ðŸ”¬ HPV vaccination coverage is SIMILAR for females and males.")
        print(f"ðŸ“Š Average coverage: Females = {female_data.mean():.1f}%, Males = {male_data.mean():.1f}%")
        print(f"ðŸ’¡ This suggests equitable vaccination programs across genders.")
else:
    print("\nâš ï¸ Insufficient data to determine gender differences.")

print("\n" + "="*70)
print("Note: This analysis focuses on HPV vaccines as they're the only")
print("vaccines in your dataset with explicit gender differentiation.")
print("Other vaccines may show different gender patterns.")
print("="*70)
```

```python
import re
def clean_vaccine_description_v2(description):
    """Alternative cleaning function that's more explicit"""
    if pd.isna(description):
        return description
    
    # Convert to string and lowercase for easier processing
    desc = str(description).lower()
    
    # Extract vaccine name
    if 'dtp' in desc:
        vaccine = 'DTP'
    elif 'polio' in desc:
        vaccine = 'Polio'
    elif 'measles' in desc:
        vaccine = 'Measles'
    elif 'bcg' in desc:
        vaccine = 'BCG'
    elif 'hepb' in desc:
        vaccine = 'HepB'
    elif 'japanese encephalitis' in desc:
        vaccine = 'Japanese encephalitis'
    elif 'malaria' in desc:
        vaccine = 'Malaria'
    else:
        # Fallback: take first word as vaccine name
        vaccine = str(description).split()[0]
    
    # Extract dose
    if '1st' in desc or 'first' in desc:
        dose = '1'
    elif '2nd' in desc or 'second' in desc:
        dose = '2'
    elif '3rd' in desc or 'third' in desc:
        dose = '3'
    elif '4th' in desc or 'fourth' in desc:
        dose = '4'
    elif 'last' in desc:
        dose = 'last'
    else:
        # Try to find any number
        num_match = re.search(r'\b(\d+)\b', desc)
        dose = num_match.group(1) if num_match else ''
    
    # Return formatted string
    if dose:
        return f"{vaccine} {dose}"
    else:
        return vaccine

# Apply the alternative function
df_clean['ANTIGEN_DESCRIPTION'] = df_clean['ANTIGEN_DESCRIPTION'].apply(clean_vaccine_description_v2)
print("\n\nUsing alternative function:")
print(df_clean)
```
