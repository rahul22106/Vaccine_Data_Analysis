```python
# converting YEAR column to numeric type and handling errors, then converting to nullable integer type
report['YEAR'] = pd.to_numeric(report['YEAR'], errors='coerce').astype('Int64')
```

```python
report.dtypes
```

```python
report['CASES'] = pd.to_numeric(report['CASES'], errors='coerce').astype('Int64')
```

```python
report.dtypes
```

```python
report.sort_values(['YEAR'], ascending=[True], inplace=True)
report.reset_index(drop=True, inplace=True)
```

```python
report.head()
```

```python
# Remove rows where GROUP contains "Created"
report = report[~report['GROUP'].str.contains('Created', na=False)]
```

```python
report = report.to_excel("processed/cleaned-reported-cases-data.xlsx", index=False)
```
