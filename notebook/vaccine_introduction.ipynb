```python
vaccine['YEAR'] = pd.to_numeric(vaccine['YEAR'], errors='coerce').astype('Int64')
```

```python
vaccine.head()
```

```python
vaccine.sort_values(['YEAR'], ascending=[True], inplace=True)
vaccine.reset_index(drop=True, inplace=True)
```

```python
vaccine.head()
```

```python
# Remove rows where GROUP contains "Created"
vaccine = vaccine[~vaccine['ISO_3_CODE'].str.contains('Created', na=False)]
```

```python
vaccine = vaccine.to_excel("processed/cleaned_vaccine_introduction_data.xlsx", index=False)
```
