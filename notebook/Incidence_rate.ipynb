```python
# sorting the dataframe by YEAR in ascending order and resetting the index
incidents.sort_values(['YEAR'], ascending=[True], inplace=True)
incidents.reset_index(drop=True, inplace=True)
```

```python
incidents.head()
```

```python
# replace null values in INCIDENCE_RATE with 0
incidents['INCIDENCE_RATE'].fillna(0, inplace=True)
```

```python
# check if the code worked
incidents.head()
```

```python
# check if there are any null values left in INCIDENCE_RATE
incidents['INCIDENCE_RATE'].isnull().sum()
```

```python
incidents['DISEASE_DESCRIPTION'].value_counts()
```

```python
# converting YEAR column to numeric type and handling errors, then converting to nullable integer type
incidents['YEAR'] = pd.to_numeric(incidents['YEAR'], errors='coerce').astype('Int64')
```

```python
# checking the data type of YEAR column
incidents['YEAR'].dtypes
```

```python
# converting DENOMINATOR values to have consistent formatting
incidents['DENOMINATOR'] = incidents['DENOMINATOR'].str.replace('per 1000 live births', 'per 1,000 live births')
```

```python
# checking if the conversion worked
incidents['DENOMINATOR'].value_counts()
```

```python
incidents['GROUP'].value_counts()
```

```python
# Remove rows where GROUP contains "Created"
incidents = incidents[~incidents['GROUP'].str.contains('Created', na=False)]
```
